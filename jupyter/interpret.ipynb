```python
import os
import yaml
import torch

from airdetect.aircraft_classification import visualize_heatmap
from wheel5.tracking import CheckpointPattern
```

```python
def snapshot_path(config, experiment, trial, epoch):
    snapshot_root = config['tracking']['snapshot_root']
    snapshot_dir = os.path.join(snapshot_root, experiment)
    return CheckpointPattern.path(os.path.join(snapshot_dir, trial), epoch)
```

```python
with open('../config.yaml', 'r') as config_file:
    config = yaml.load(config_file, Loader=yaml.Loader)
    
path = snapshot_path(config, 
                     experiment='resnet50_mix2', 
                     trial='lrA=3.27e-04, lrB=1.96e-04, lr_f=2.0, lr_t0=10.0, lr_w=3.0, nn_frz=4.0, wdA=0.0248, wdB=0.1592, x_cut=0.0, x_cut_a=0.3012, x_lbs=1.80e-04, x_mxp=0.0, x_mxp_a=0.6812-2020-03-31_13:17:39.290138', 
                     epoch=69)
    
visualize_heatmap(dataset_config=config['datasets']['train'],
                  snapshot_path=path,
                  device=1,
                  samples=10,
                  inter_mode='bilinear',
                  cutoff_ratio=0.12);
```

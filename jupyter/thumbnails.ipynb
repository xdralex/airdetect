```python
import os
import pathlib
import numpy as np
import pandas as pd
import yaml

from PIL import Image
from wheel5.transforms import SquarePaddedResize
```

```python
def convert_thumbnails(transform, original_dir, thumbnail_dir):
    pathlib.Path(thumbnail_dir).mkdir(parents=True, exist_ok=False)
    
    for filename in os.listdir(original_dir):
        print(filename)
        
        original_path = os.path.join(original_dir, filename)
        original_img = Image.open(original_path)
        
        thumbnail_path = os.path.join(thumbnail_dir, filename)
        thumbnail_img = transform(original_img)
        thumbnail_img.save(thumbnail_path)
```

```python
with open('../config.yaml', 'r') as config_file:
    config = yaml.load(config_file, Loader=yaml.Loader)
```

```python
transform = SquarePaddedResize(size=224, interpolation=Image.LANCZOS)
```

```python
convert_thumbnails(transform, 
                   original_dir=config['datasets']['train']['image_dir'], 
                   thumbnail_dir='/data/ssd/datasets/airliners/thumbnails/train')
```

```python
convert_thumbnails(transform, 
                   original_dir=config['datasets']['public_test']['image_dir'], 
                   thumbnail_dir='/data/ssd/datasets/airliners/thumbnails/public_test')
```

```python
convert_thumbnails(transform, 
                   original_dir=config['datasets']['private_test']['image_dir'], 
                   thumbnail_dir='/data/ssd/datasets/airliners/thumbnails/private_test')
```

```python
import os
from pathlib import Path

from PIL import Image
from PIL import ExifTags
```

```python
orig_dir = '/data/ssd/datasets/airdetect/reallife_orig'
image_dir = '/data/ssd/datasets/airdetect/reallife'

Path(image_dir).mkdir(parents=True, exist_ok=False)

for filename in os.listdir(orig_dir):
    image = Image.open(os.path.join(orig_dir, filename))
    
    exif=dict((ExifTags.TAGS[k], v) for k, v in image._getexif().items() if k in ExifTags.TAGS)
    orientation = exif.get('Orientation') or 1

    if orientation == 1:
        pass
    elif orientation == 2:
        image = image.mirror()
    elif orientation == 3:
        image = image.rotate(180, expand=True)
    elif orientation == 4:
        image = image.rotate(180, expand=True)
        image = image.mirror()
    elif orientation == 5:
        image = image.rotate(270, expand=True)
        image = image.mirror()
    elif orientation == 6:
        image = image.rotate(270, expand=True)
    elif orientation == 7:
        image = image.rotate(90, expand=True)
        image = image.mirror()
    elif orientation == 8:
        image = image.rotate(90, expand=True)
    else:
        raise AssertionError(f'Unsupported orientation {orientation}')
    
    image.save(os.path.join(image_dir, filename))
    print(f'processed: {filename}')
```
